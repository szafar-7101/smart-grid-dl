```python
import pandas as pd
import numpy as np 

pd.set_option('display.max_columns', None)

path = "../data/dataset.csv"

df = pd.read_csv(path)
print(df.shape)
df.head()
```

```python
print(df.columns)
print(df.dtypes)
print(df.info())
```

```python
### Renaming columns 
column_map = {
    "datetime": "timestamp",
    "nat_demand": "load",
    "T2M_toc": "temperature",
    "QV2M_toc": "humidity",
    "WS10M_toc": "wind_speed",
    "PRECTOTCORR_toc": "precipitation"
}

for old, new in column_map.items():
    if old in df.columns:
        df = df.rename(columns={old: new})
```

```python
df.head()
```

```python
# Convert TImestamp and Sort 

df['timestamp'] = pd.to_datetime(df['timestamp'])
df = df.sort_values('timestamp')
df = df.reset_index(drop=True)
```

```python
print(df['timestamp'].min())
print(df['timestamp'].max())
```

```python
"""
The dataset spans from 
2015-01-03 01:00:00
2020-06-27 00:00:00
"""
```

```python
### Check for duplicate timestamps 
duplicates = df['timestamp'].duplicated().sum()
print(f"Number of duplicate timestamps: {duplicates}")
```

```python
# No duplicate timestamps found.

# Next important thing is to check for Missing time gaps 

expected = pd.date_range(
    start = df['timestamp'].min(),
    end = df['timestamp'].max(),
    freq = 'h'
)

missing_times = expected.difference(df['timestamp'])

print(f"Missing timestamps: {len(missing_times)} ")
missing_times[:10]
```

```python
# There were no missing timestamps found either

missing = df.isnull().sum().sort_values(ascending=False)
missing_percent = (df.isnull().mean() * 100).sort_values(ascending=False)

pd.DataFrame({
    "missing_count": missing,
    "missing_percent": missing_percent
})
```

```python
# There are no missing values in the dataset. We can move on normally now 

# lets convert the dataset to paraquet for faster loading in the future
df.to_parquet("../data/dataset.parquet", index=False)
```
